```python
titanic_df['Alone'] = titanic_df['Alone'].apply(lambda lone: 0 if lone == 'Alone' else 1)
```

```python
titanic_df['Person'] = titanic_df['Person'].apply(lambda persn: 0 if persn == 'male' else 1 if persn == 'female' else 2)
```

```python
titanic_df.head(10)
```

```python
titanic_df['Embarked2'] = titanic_df['Embarked'].fillna('Other')
titanic_df['Embarked2'] = titanic_df['Embarked2'].apply(lambda em: 1 if em == 'S' else 2 if em == 'C' else 3 if em == 'Q' else 0)
titanic_df.head()
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('N',inplace = True)
```

```python
titanic_df['Age2'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
```

```python
import re
```

```python
def abbrev(abb):
    a = re.findall('Mr.|Mrs.|Miss.|Master.',abb)
    for item in a:
        return item
```

```python
titanic_df['name abb'] = titanic_df['Name'].apply(lambda abb: abbrev(abb))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
titanic_df['name abb'] = titanic_df['name abb'].astype('str')
```

```python
titanic_df['name abb'] = titanic_df['name abb'].fillna('other').astype('category')
titanic_df['name abb'] = le.fit_transform(titanic_df['name abb'])
```

```python
titanic_df['Sex'] = titanic_df['Sex'].apply(lambda sex: 0 if sex == 'male' else 1)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
train,test = train_test_split(titanic_df, test_size = 0.3)
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
et = RandomForestClassifier(n_estimators=1000, max_depth=None, min_samples_split=2, random_state=0,verbose =1)
columns = ["Fare","Person","Alone","Pclass","name abb","Age2","Embarked2","Sex"]#,"Age2","name abb"]
labels = train["Survived"].values
features = train[list(columns)].values
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp.fit(features)
predictions = et.fit(features,labels).predict(imp.transform(test[columns].values))
test['Survived_pred'] = predictions
scr = et.score(features,labels)
print(scr)
```

```python
from pandas_ml import ConfusionMatrix
c = ConfusionMatrix(test['Survived'],test['Survived_pred'])
c
```

```python
c.print_stats()
```

```python
c.plot(backend='seaborn')
```

```python

importances = et.feature_importances_
std = np.std([tree.feature_importances_ for tree in et.estimators_],axis=0)
indices = np.argsort(importances)[::-1]
print("Feature ranking:")

for f in range(features.shape[1]):
    print("%d. feature %d (%f ->" % (f + 1,indices[f], importances[indices[f]])+" "+columns[int(indices[f])]+")")
# Plot the feature importance of the forest
plt.figure(figsize=(13,8))
plt.title('Feature Importance')
plt.bar(range(features.shape[1]), importances[indices], color="blue", yerr=std[indices], align="center")
plt.xticks(range(features.shape[1]), indices)
plt.xlim([-1, features.shape[1]])
```

```python
arr = np.column_stack((columns,list(et.feature_importances_)))
dfplt2 = pd.DataFrame(arr)
dfplt2[1] = dfplt2[1].astype('float')
dfplt2.columns = ['Feature','Value']
dfplt2 = dfplt2.sort_values('Value',ascending=False)
dfplt2.plot(kind='bar',figsize=(10,8))
dfplt2
```
